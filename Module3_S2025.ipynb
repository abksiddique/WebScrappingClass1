```python
weather= requests.get("https://forecast.weather.gov/MapClick.php?lat=39.12711000000007&lon=-84.51438999999993")
```

```python
soup= BeautifulSoup(weather.content, 'html.parser')
```

```python
soup.prettify()
```

```python
sven_day= soup.find(id='seven-day-forecast-container')
```

```python
print(sven_day)
```

```python
forecast= sven_day.find(class_='forecast-tombstone')
forecast.prettify()
```

```python
period= forecast.find(class_='period-name').get_text()
temp= forecast.find(class_='temp').get_text()
s_desc=forecast.find(class_='short-desc').get_text()
img=forecast.find('img')
#img=forecast.find(class_=forecast-icon)
desc=img['title']
print(period,temp,s_desc,desc)
```
